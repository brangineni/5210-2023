## #04 -

Which of those State values has the best average performance on the `Star Rating for health team communicated well with them` score?

```python
average_star_rating = data.groupby('State')['Star Rating for health team communicated well with them'].mean()

# Find the State with the best average performance
best_state = average_star_rating.idxmax()
best_avg_rating = average_star_rating.max()

print(f'The State with the best average performance is "{best_state}" with an average rating of {best_avg_rating:.2f}.')
```
